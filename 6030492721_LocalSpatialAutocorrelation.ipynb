```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

## Load data 

```python
df=pd.read_excel('selected_subdistrict_elder_data.xlsx')
df_contiguity=pd.read_excel('contiguitymatrix.xlsx')
```

```python
df
```

```python
df_contiguity
```

```python
df_contiguity['sum'] = df_contiguity.sum(axis = 1)
```

```python
df_contiguity['ประเวศ']=df_contiguity['ประเวศ']/df_contiguity['sum']
df_contiguity['ทับช้าง']=df_contiguity['ทับช้าง']/df_contiguity['sum']
df_contiguity['คลองสองต้น่น']=df_contiguity['คลองสองต้น่น']/df_contiguity['sum']
df_contiguity['ลาดกระบัง']=df_contiguity['ลาดกระบัง']/df_contiguity['sum']
df_contiguity['ดอกไม้']=df_contiguity['ดอกไม้']/df_contiguity['sum']
df_contiguity['หนองบอน']=df_contiguity['หนองบอน']/df_contiguity['sum']
df_contiguity['อ่อนนุช']=df_contiguity['อ่อนนุช']/df_contiguity['sum']
df_contiguity['พัฒนาการ']=df_contiguity['พัฒนาการ']/df_contiguity['sum']
```

```python
df_contiguity
```

```python
z_score=list(stats.zscore(df['g_total']))
list_subdist=list(df_contiguity.columns[1:9])
for i in range(8):
    df_contiguity[list_subdist[i]]=df_contiguity.loc[:,list_subdist[i]]*z_score[i]
```

```python
df_contiguity
```

```python
df_contiguity['sum_Wij_Zj'] = df_contiguity[df_contiguity.columns[1:9]].sum(axis = 1)
```

```python
df_contiguity
```

```python
df_contiguity['LISA']=0
for i in range(8):
    df_contiguity['LISA'].iloc[i]=df_contiguity['sum_Wij_Zj'].iloc[i]*z_score[i]
```
